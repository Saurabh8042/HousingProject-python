**6.**Write a programming construct (create a user defined function)
to calculate the median value of the data set wherever required.

```python
# Function for calculating median


def findMedian(a, n):

    # First we sort the array
    sorted(a)

    # check for even case
    if n % 2 != 0:
        return float(a[int(n/2)])

    return float((a[int((n-1)/2)] +
                  a[int(n/2)])/2)

findMedian(data['longitude'],2084)
```

**10.**Please create a new column named total_bedroom_size.
If the total bedrooms is 10 or less, it should be quoted as small.
If the total bedrooms is 11 or more but less than 1000, it should be medium,
otherwise it should be considered large.


```python
total_bedroom_size=[]
for room in data['total_bedrooms']:
  if room<=10:                 total_bedroom_size.append(['small'])
  elif room>10 and room<=1000: total_bedroom_size.append(['medium'])
  else:                        total_bedroom_size.append(['large'])

data['total_bedroom_size']=total_bedroom_size
data['total_bedroom_size']
```
